```python
#MATERIAL DO CURSO

#1. Com base na equação geral da reta Ax+By+C, onde A = 3, B = -2 e C = -6
#TRAÇAR RETA COM A FUNÇÃO AX+BY+C -> 3X + -2Y + -6
a = 3 				#o A vale 3
b = -2 				#o B vale -2
c = -6 				#o C vale -6
x =float (input('x = '))
y =float (input('y = '))
resultado = (a*x) + (b*y) + c
print(resultado)
Ex. 	x = 20
y = 30
-6.0

#2. Ano bissexto
#PERGUNTAR SE O ANO É BISSEXTO E INFORMAR TRUE OU FALSO
ano = int(input('Ano: '))
bissexto = (ano%4==0) and (ano%100!=0) or (ano%400==0)
print ('O ano', ano, 'é bissexto?', bissexto)
Ex. 	O ano 2020 é bissexto? True

ano = int(input('Ano: '))
if (ano % 400 == 0):
  print('É bissexto')
elif (ano % 100 == 0):
  print ('Não é bissexto')
elif (ano % 4 == 0):
  print ('É bissexto')
else:
  print('Não é bissexto')
print ('Fim!')
Ex. Ano: 2021
Não é bissexto
Fim!

#3. NDVI - Exemplos
ndvi = float(input('NDVI: '))
if (ndvi > 0.3) and (ndvi < 0.8):
  print('Vegetação densa')
print ('NDVI: ', ndvi)
Ex. 	NDVI: 0.5
vegetação densa
NDVI:  0.5

ndvi = float(input('NDVI: '))
if (ndvi > 0.3) and (ndvi < 0.8):
  print('Vegetação Densa!')
else:
  print('Pouca Vegetação!')
print('NDVI: ', ndvi)
Ex. 	NDVI: 0.3
Pouca Vegetação!
NDVI:  0.3

ndvi = float(input('NDVI: '))
if (ndvi < -1) or (ndvi > 1):
  print ('NDVI fora do intervalo!')
elif (ndvi <0):
  print ('Valor negativo')
else:
  print ('Valor positivo')
  if (ndvi > 0.3) and (ndvi < 0.8):
    print('Vegetação Densa!')
  else:
      print('Pouca Vegetação')
print ('NDVI:', ndvi)
Ex. NDVI: 0.7
Valor positivo
Vegetação Densa!
NDVI: 0.7

#4. Exemplo de gráfico com o matplotlib (gráfico feito para o trabalho de Princípios Físicos)
import matplotlib.pyplot
bandas = ['Azul', 'Verde', 'Vermeho']
geometria1 = [5, 7, 13]
geometria2 = [6, 9, 14]
geometria3 = [5, 7, 12]
geometria4 = [6, 9, 14]
media1 = [5, 8, 13]
matplotlib.pyplot.plot(bandas, geometria1, label = 'Geometria 1', color = 'blue')
matplotlib.pyplot.plot(bandas, geometria2, label = 'Geometria 2', color = 'green')
matplotlib.pyplot.plot(bandas, geometria3, label = 'Geometria 3', color = 'yellow')
matplotlib.pyplot.plot(bandas, geometria4, label = 'Geometria 4', color = 'red')
matplotlib.pyplot.plot(bandas, media1, label = 'Média', color = 'black')
matplotlib.pyplot.title ('Solo Exposto (arenoso)', fontsize = 15)
matplotlib.pyplot.xlabel ('Bandas', fontsize = 15)
matplotlib.pyplot.ylabel ('Albedo (%)', fontsize=13)
matplotlib.pyplot.legend()
from google.colab import files
matplotlib.pyplot.savefig("soloexpostoabedo.png", dpi=300)
files.download("soloexpostoabedo.png")

 
#5. Script feito para o trabalho de Princípios Físicos (Irradiação Solar e Irradiância)
import math

#CONSTANTE SOLAR EM UMA HORA
constantesolar1hr = float(1.367 * 3.6)

#DISTÂNCIA TERRA-SOL NO DIA DE INTERESSE
diajuliano = float(input('Informe o dia (fev c/ 28 dias, ou seja, de 1 a 365): '))
angulodia = float(math.pi * 2 * (diajuliano - 1) / 365) 
distanciaterrasol = float(1.000110 + 0.034221 * math.cos(angulodia) + 0.001280 * math.sin(angulodia) + 0.000719 * math.cos(2*angulodia) + 0.000077 * math.sin(2*angulodia))
constantesolardia = (constantesolar1hr * distanciaterrasol)

#IRRADIAÇÃO SOLAR
declinacaosolar = math.radians(float(input('Declinação solar do dia de interesse em º (ver Tabela 2.2 de Lorenzetti (2015)): ')))
latitude = math.radians(float(input('Latitude da região de interesse em º: ')))
horario = float(input('Horário de interesse (p. ex: 10:30h -> 10.5; 10:15 -> 10.25...): '))
angulohorariosolar = math.radians(float(horario-12)) * 15 
angulohorariosolargrau = math.degrees(angulohorariosolar)

superficiehorizontal = float(math.sin(declinacaosolar) * math.sin(latitude) + math.cos(declinacaosolar) * math.cos(latitude) * math.cos(angulohorariosolar)) 
irradiacaosolar = (constantesolar1hr * distanciaterrasol) * superficiehorizontal

#IRRADIÂNCIA
irradiancia = float(irradiacaosolar / 3.6)

#EXIBINDO OS RESULTADOS
print ("""
""" 'Constante solar: ', constantesolar1hr, 'kJ m-² h-¹')
print ('Constante solar corrigida para o dia de interesse: ', constantesolardia, 'kJ m-² h-¹')
print ('Distância Terra-Sol no dia de interesse: ', distanciaterrasol, 'UA')
print ('Ângulo solar no horário de interesse: ', angulohorariosolargrau, 'graus')
print ("""
""" 'Irradiação Solar no Topo da Atmosfera: ', irradiacaosolar, 'kJ m-² h-¹')
print ('Irradiância: ', irradiancia, 'KJ m-2 ')

 
#6. Exemplo de formatação com (f’ {}’)
print (f'texto... {variavel}')

#7. Laços do tipo for
for num in range(10):
    if (num == 5):
        break
    print(f'Número: {num}')
print("Primeiro comando fora do laço!")

menor = int(input('Digite o menor valor do intervalo: '))
maior = int(input('Digite o maior valor do intervalo: '))
for x in range (menor, maior + 1):
    if (x % 2):
        continue
    print(x)

def FatorialFor (number):
  resultado = 1
  for i in range (1, number+1):
    resultado = resultado * i
    print(resultado)
number = int(input('Fatorial de: '))
print (FatorialFor(number))

for i in range (0, 320, 20): #percorre 0 a 320 de 20 em 20
  print(i) 

for linha in range (3):
  for coluna in range(3):
    for banda in range (3):
      print(banda, coluna, linha)

soma = 0
for i in range(10, 51):
  soma = soma + i
print (f"soma: {soma}")

n = int(input('Número: '))
print (n)
if n>10 or n<1:
  print('Número resultante: ')
else:
  fatorial = 1
  for a in range (1, n +1):
    fatorial = fatorial * a
  print (fatorial)

x = int(input('x: '))
y = int(input('y: '))
potencia=1
if y >= 0:
  for i in range (1, y+1):
    potencia=potencia*x
  print ('Resultado:', potencia)
  print ('Prova Real: ', x ** y)
  print ('Deu certo: ', potencia == (x ** y))
else:
  print('y fora do intervalo: ')

#8. Laços do tipo while
def FatorialWhile (number1):
  resultado = 1
  count = 1
  while (count <= number1):
    resultado = resultado * count
    count = count + 1
    print(resultado)
number1 = int(input("Fatorial de: "))
print (FatorialWhile(number1))

def Fatorial1(num):
  if (num < 0) or (type(num) != int):
    raise ValueError("O Fatorial só é definido para inteiros positivos!")
  else:
      produto = 1
      while(num > 0):
        produto = produto * num
        num = num - 1
      return produto
fat = int(input('Fatorial de: '))
print (Fatorial1(fat))

i = 1
soma = 0
while i <=5:
  soma = soma+i
  i = i+1
print('Soma Final:', soma)
print ('Valor final de i:', i)

tmin = 0
tmax = 300
delta_t = 20
fahr = tmin
while fahr <= tmax:
  celsius = 5*(fahr-32) / 9
  print (fahr, celsius)
  fahr = fahr+delta_t

for fahr in range (0, 320, 20):
  celsius=5*(fahr-32)/9
  print (fahr, celsius)

9 Tabuada
for linha in range (1, 11):
  for coluna in range(1,11):
    resultado = linha*coluna
    print (linha, 'x', coluna, '=', resultado)

#9. Matrizes

#9.1 Exemplo incerto que gera matrizes
def gerar_matriz (n_linhas, n_colunas):
    return [[""]*n_linhas for _ in range(n_colunas)]
n_linhas = int(input('Informe a quantidade de linhas: '))
n_colunas = int(input('Informe a quantidade de colunas: '))
print (gerar_matriz(n_linhas, n_colunas))

#9.2 Criando listas para matrizes
def main():
    a = [0, 1, 2, 3, 4]
    b = a[:]
    c = a[:]
    d = a[:]
    e = a[:]
    print("a = ", a)
    print("b = ", b)
    print("c = ", c)
    print("d = ", d)
    print("e = ", e)
main()

#10. Dicionários

#10.1 Dicionários – Exemplo 1
sao_paulo = {
    'name' : 'São Paulo',
    'woeid' : 12582314,
    'bounding-box' : ((-46.82, -24.00), (-46.36, -23.68)),
    'country' : 'Brazil'
}
print (sao_paulo)
type (sao_paulo)

#10.2 Dicionários – Exemplo 2
sjc = {
    'longitude': -45.88,
    'latitude': -23.17
}
ouro_preto = {
    'longitude': -43.50,
    'latitude': -20.38
}
print(sjc)
print(ouro_preto)
longitude = sjc['longitude']
print (longitude)
latitude = sjc.get('latitude')
print(latitude)
sjc ['latitude'] = -20.379
del sjc ['latitude']
list(sjc)
list(sorted(sjc))

#10.3 Dicionários – Exemplo 3
dict([("xmin", -46.8254), ("ymin", -24.0084), ("xmax", -46.3648), ("ymax", -23.6830)])
{'xmin': -46.8254, 'ymin': -24.0084, 'xmax': -46.3648, 'ymax': -23.683}

#10.4 Dicionários – Exemplo 4
dictionary = dict( longitude=-45.88, latitude=-23.17)
print (dictionary)
dict1 = {'longitude': -45.88, 'latitude': -23.17}
print (dict1)
d={}

#10.5 Dicionários – Exemplo 5 – Operação Matemática
xs = [1, 2, 3, 4, 5]
f = {x: (1.5 * x - 3) for x in xs}
print(f)
print(f[3])
type (f)

#10.6 Dicionários – Exemplo 6 – Conjuntos
mg = {'Mariana', 'Ouro Preto', 'Ouro Branco'}
rn = {'Ouro Branco', 'Acari', 'Caicó', 'Cruzeta'}
type (mg)
print (mg & rn) #intersecção
print (mg - rn) #exclusivo
print (mg | rn) #união
print (mg ^ rn) #diferença simétrica

#10.7 Formas de fazer dicionários
d = {} #dicionário vazio
e = dict() #dicionário vazio
g = set() #conjunto vazio
set () #conjunto vazio
frozenset () #conjunto imutável

#11. Formatação de strings
curso = 'CAP-419/SER-347'
'Bem vindo a %s!'

salario = 1250.34
'Salario: R$ %f' % salario
'Salario: R$ %.2f' % salario

'%05d' % 30 #padding

plataforma = 'Landsat-8'
sensor = 'OLI'
ano = 2013
'%s foi lançado em %d e possui o sensor %s' % (plataforma, ano, sensor)

idade = 25
'Idade: {}'.format(idade)

salario = 1250.34
print ("Salario: R$ {}".format(salario))
print ("Salario: R$ {:.2f}".format(salario))

nome = 'Alisson de Oliveira'

for letra in nome:
  print (letra)

for i in range (len(nome)):
  print (nome[i])

cidades = ["São Paulo", "Rio de Janeiro", "Belo Horizonte"]
for cidade in cidades:
    print("Nome:", cidade)

#11.1 Atravessando duas listas e convertendo Fahreheint em Celsius
print('Conversão de F -> C')

fahr = [ 0, 20, 40, 60, 80, 100 ]
celsius = [ 5*(x-32)/9 for x in fahr ]

for f, c in zip(fahr, celsius):
    print(f, c)

print('Fim!')

#11.2 Atravessando três listas e convertendo Fahrenheit em Celsius e em Kelvin

print('Conversão de F -> C -> K')
fahr = [ 0, 20, 40, 60, 80, 100 ]
celsius = [ 5*(x-32)/9 for x in fahr ]
kelvin = [ x+273 for x in celsius ]
for f, c, k in zip(fahr, celsius, kelvin):
    print(f, c, k)
print('Fim!')

#12. STRINGS
s = '10'
t = '5'
s+t
4 * s

#12.1 PERTINÊNCIA
'.tif' in 'imagem-cbers.tif'

#12.2 IMPERTINÊNCIA
'cbers' not in 'imagem-cbers.tif'

#12.3 CAPITALIZE
nome_da_imagem = 'cbers.tif'
nome_da_imagem.capitalize()

#13. DEFININDO FUNÇÕES
#Parâmetros Default
def graustoradianos(alpha, pi=3.14):
  x = alpha * pi / 180
  return x
resultado1 = graustoradianos(45)
print (resultado1)

#Parâmetros Default
def graustoradianos(alpha, pi=3.14):
  x = alpha * pi / 180
  return x
resultado3 = graustoradianos(pi=3.141516, alpha=45)
print (resultado3)

def myprint(*args):
  for a in args:
    print (a)
myprint('p1', 'p2')

def myprint(*args):
  for a in args:
    print (a)
myprint(1, 3, 5, 7, 9)

def myprint(*args):
  for a in args:
    print (a)
myprint(-20.38, -43.50, 'Ouro Preto')

def myprint(**kwargs):
  for k, v in kwargs.items():
    print ('Parâmetro: {}, valor: {}'.format(k, v))
myprint (latitude=-20.38, longitude=-43.50, Nome='Ouro Preto')

def myprint1(alpha, pi=3.14):
  x=alpha*pi/180
  return x
def myprint3(*args):...
def myprint4(**kwargs):...
l = [45, 3.1]
d={
    'pi':3,
   'alpha':45
}
graustoradianos(*l)


pairs = [(1, 'one'), (2, 'two'), (3, 'three'), (4, 'four')]
pairs.sort(key=lambda pair:pair[1])
print(pairs)

pairs = [(1, 'one'), (2, 'two'), (3, 'three'), (4, 'four')]
pairs.sort(key=lambda pair:pair[0]) #ordenado pelo primeiro elemento
print(pairs)

l = [1,2,3,4,5]
pot = map(lambda x: x**2, l)
print (list(pot))

def quadrado(x):
  return x ** 2

l = [1,2,3,4,5]
pot = map(quadrado, l)
print (list(pot))

u = [1,2,3,4,5]
v = [10,11,12,13,14]
soma = map(lambda x, y: x+y, u, v)
valores = list(soma)
print (valores)

ESCOPO DE VARIÁVEIS
print (locals().keys())
print (globals().keys())

Prof. Gilberto - Simulado – Exercício – Construir um jogo
import getpass

def ocorre(letra, segredo):
    return letra in segredo
def ocultar(letras, palavra_secreta):
    palavra_oculta = ''
    for c in palavra_secreta:
        if c in letras:
            palavra_oculta = palavra_oculta + c
        else:
            palavra_oculta = palavra_oculta + '_'
    return palavra_oculta

palavra_secreta = getpass.getpass(prompt='Jogador 1, por favor, entre com uma palavra secreta: ')
chances = 6
letras_ja_digitadas = []
texto_oculto = ocultar(letras_ja_digitadas, palavra_secreta)

while (chances > 0) and (texto_oculto != palavra_secreta):
    print('Número de chances:', chances)
    l = input('Jogador 2, adivinhe uma letra da palavra secreta: ')
    if l in letras_ja_digitadas:
        texto_oculto = ocultar(letras_ja_digitadas, palavra_secreta)
        print('A letra "{}" já foi digitada! {}'.format(l, texto_oculto))
        chances = chances - 1
    elif not ocorre(l, palavra_secreta):
        letras_ja_digitadas.append(l)
        texto_oculto = ocultar(letras_ja_digitadas, palavra_secreta)
        print('A letra "{}" não ocorre na palavra secreta! {}'.format(l, texto_oculto))
        chances = chances - 1
    else: # Caso a letra não tenha sido digitada previamente e faça parte da palavra secreta
        letras_ja_digitadas.append(l)
        texto_oculto = ocultar(letras_ja_digitadas, palavra_secreta)
        print('A letra "{}" ocorre na palavra secreta! {}'.format(l, texto_oculto))
    print('---------')
    print()
if chances > 0:
    print('Parabéns! Você descobriu a palavra secreta:', palavra_secreta)
else:
    print('Fim do jogo! Você não descobriu a palavra secreta:', palavra_secreta)
```
